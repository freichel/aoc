```python
filename = "2021-02.txt"
commands = [(line.strip().split(" ")[0], -int(line.strip().split(" ")[1]) if line.strip().split(" ")[0] == "up" else int(line.strip().split(" ")[1])) for line in open(filename).readlines()]
commands
```

```python
hor = 0
dep = 0
for cmd, val in commands:
    if cmd == "forward":
        hor += val
    else:
        dep += val
hor*dep
```

```python
hor = 0
dep = 0
aim = 0
for cmd, val in commands:
    if cmd == "forward":
        hor += val
        dep += aim * val
    else:
        aim += val
hor*dep
```
